```python
system_data = [
{"TransactionID": "T001", "Date": "2025-08-20", "Description": "Vendor A", "Amount":500},
{"TransactionID": "T002", "Date": "2025-08-21", "Description": "Vendor B", "Amount": 400},
{"TransactionID": "T003", "Date": "2025-08-22", "Description": "Vendor C", "Amount": 450},
{"TransactionID": "T004", "Date": "2025-08-23", "Description": "Vendor D", "Amount": 600}
]
```

```python
bank_data = [
    {"TransactionID": "T001", "Date": "2025-08-21", "Description": "Vendor A", "Amount": 500},
    {"TransactionID": "T002", "Date": "2025-08-22", "Description": "Vendor B", "Amount": 400},
    {"TransactionID": "T005", "Date": "2025-08-22", "Description": "Vendor E", "Amount": 800}
]
```

```python
import pandas as pd
```

```python
df_system = pd.DataFrame(system_data)
df_system
```

```python
df_bank = pd.DataFrame(bank_data)
df_bank
```

```python
# Merging the two dataset on TransactionID with Outer join to capture the mismatch from both the sides

reconciliation = pd.merge(df_system, df_bank, on = 'TransactionID', how = 'outer', suffixes = ('system', 'bank'))
reconciliation
```

```python
# Identify mismatch

def status(row):
  if pd.isnull(row["Amountsystem"]):
    return "mismatch"
  if pd.isnull(row["Amountbank"]):
    return "mismatch"
  if row["Amountsystem"] != row["Amountbank"]:
    return "mismatch"
  else:
    return "match"
```

```python
reconciliation['status'] = reconciliation.apply(status, axis = 1)
reconciliation
```
